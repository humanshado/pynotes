```python
# To sum x and y, we need to write a loop
result = []
for x, y in zip(x, y):
    result.append(x + y)

result
```

Matplotlib Primer 

```python
x = np.linspace(0,2*np.pi, 50)
plt.plot(np.sin(x))
```

```python
#multiple data sets
plt.plot(x, np.sin(x), x, np.sin(2*x))
```

```python
#multiple plots on same figure
plt.subplot(2,2,1)
plt.plot(np.sin(x), label='sin x')
plt.legend(loc='best')
plt.axis('tight')
plt.subplot(2,2,2)
plt.plot(np.cos(x), 'r', label='cos x')
plt.legend(loc='best')
plt.axis('tight')
plt.subplot(2,2,3)
plt.plot(np.sin(2*x), 'g', label='sin2x')
plt.legend(loc='best')
plt.axis('tight')
plt.subplot(2,2,4)
plt.plot(np.cos(2*x), 'b', label='cos 2x')
plt.legend(loc='best')
plt.axis('tight')
```
